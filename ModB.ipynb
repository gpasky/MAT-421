**17.1: Interpolation Problem Statement**



*   Interpolation is a technique where an estimation function is used to traverse through existing and reliable data points

*   Regression uses estimates of data points and works to minimize measurement errors between the data and model

Estimation function used for interpolation: $\hat{y}(x_i) = y_i$





```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [2, 3.2, 5, 6]
y = [1, 3, 4, 7]

plt.figure(figsize = (4,2))
plt.plot(x, y, 'ob')
plt.title('Interpolation Data')
plt.ylim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.2 Linear Interpolation**



*   Linear Interpolation is the estimated point on the line closest to its left and right neighbors

*   Linear Interpolation at x: $\hat{y}(x)$ = $y_i$ + $\frac{(y_{i+1}-y_i)(x-x_i)}{x_{i+1}-x_i}$



```python
# An example of Linear Interpolation with data can be made
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [4, 7, 9]
y = [1, 6, 10]

func = interp1d(x,y)
y_hat = func(4.5)
print(y_hat)


```

This can also be solved by plugging in the values to the equation using the first and second data points with x = 4.5 where 1 < x < 6:

$\hat{y}(x)$ = $1$ + $\frac{(6-1)(4.5-4)}{(7-4)}$ = 1.83


```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(4.5, y_hat, 'ro')
plt.title('Linear Interpolation of data (x = 4.5)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.3 Cubic Spline Interpolation**



*   The Cubic Spline Interpolation involves a set of piecewise cubic functions

*   Cubic Polynomial: $S_i(x_i) = a_ix^3 + b_ix^2 + c_ix + d_i$

*   We need 4(n-1) equations to find and determine the coefficients $a_i, b_i, c_i, d_i$

Example: Cubic spline interpolation of the data set x = [0,1,2,3] and y = [1,3,2,4] for $0 \leq x \leq 3$





```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0,1,2,3]
y = [1,3,2,4]

func = CubicSpline(x,y,bc_type='natural')
x_new = np.linspace(0,3,100)
y_new = func(x_new)

plt.figure(figsize=(10,8))
plt.plot(x_new, y_new,'b')
plt.plot(x,y,'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
