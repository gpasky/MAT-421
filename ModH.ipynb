**22.1 ODE Initial value Problem Statement**

Ordinary Differential Equation(ODE): a differential equation where the independent variable and its derivatives is in 1D.

$n^{th}$ order ODE:
$F(x,f(x),\frac{df(x)}{dx},\frac{d^2f(x)}{d^2x},\frac{df^3(x)}{dx^3},...,\frac{d^{n}f
(x)}{dx^{n-1}} = \frac{d^{n}f(x)}{dx^n}$

Partial Differential Equation(PDE): a form of a differential equation where x is a vector with independent variables $x_1,x_2,x_3,...,x_m$ for any order or combination of variables:

$\frac{\partial u(t,x,y,z)}{\partial t} = Î±(\frac{\partial u(t,x,y,z)}{\partial x}+\frac{\partial u(t,x,y,z)}{\partial y}+\frac{\partial u(t,x,y,z)}{\partial z})$

Particular Solution: a p(x) that satisfies the differential equation and $p^{(j)}(x_i)=p_i$, where $p^{(j)}$ is the $j^{th}$ derivative of $p$, for $n$ triplets, $(j,x_i,p_i)$

Example: Simple Harmonic Motion

Restoring force: $F = -kx(t)$

Newton's 2nd Law: $F=m\frac{d^2x(t)}{dt^2}$

$-kx(t)=m\frac{d^2x(t)}{dt^2}$

Then,
$\frac{d^2x(t)}{dt^2}+\frac{k}{m}x(t)=0$

General solution:
$x(t)=Acos(\sqrt{\frac{k}{m}}t)+Bsin(\sqrt{\frac{k}{m}}t)$

First derivative:

$\frac{dx(t)}{dt}=-\sqrt{\frac{k}{m}}sin(\sqrt\frac{k}{m}t)$

Second derivative:

$\frac{dx(t)}{dt}=-\sqrt{\frac{k}{m}}cos(\sqrt\frac{k}{m}t)$

Plug second derivative into the differential equation:

$m(-\sqrt{\frac{k}{m}}cos(\sqrt\frac{k}{m}t)) = -kcos(\sqrt{\frac{k}{m}}t)$

Thus,

$-kcos(\sqrt{\frac{k}{m}}t)=-kcos(\sqrt{\frac{k}{m}}t)$

Particular Solution:

$x_0$ coefficient will be used for derivatives to satisfy the equation.

$x(0)=x_0cos(0) = x_0$, and $0=x_0\sqrt{\frac{k}{m}}sin(0)=0$

Thus,

The particular solution satisfies the known values.


**22.2 Reduction of Order**

We can solve initial value problems by reducing the order of the differential equation to first order. This is done using $S(t)$, the state of the system as a function of time.

State of the system: collection of all the dependent variables that are relevant to the behavior of the system.

$f^{(n)}=F(t,f(t),f^{1}(t),f^{2}(t),f^{3}(t),...,f^{n-1}(t))$

The derivative of the state of the system is taken where $S_i(t)$ is the $i^{th}$ element of $S(t)$. This turns an $n^{th}$ order ODE into n first order ODEs with the same terms.

Example: A change in population of fish, f(t), and bears, b(t):

$\frac{df(t)}{dt}=4f(t)-2b(t)$
and
$\frac{db(t)}{dt}=f(t)-b(t)$

$S(t)=\begin{bmatrix} f(t)\\ b(t)\end{bmatrix}$

We then get the first-order ODE equivalent:

$\frac{dS(t)}{dt}\begin{bmatrix} 4 & -2\\ 1 & 1\end{bmatrix}$

**22.3 The Euler Method**

Explicit Euler Formula:
$S(t_{j+1})=S(t_j)+hF(t_j,S(t_j))$

*   Allows for approximation of the state at $S(t_{j+1})$ given the state at $S(t_j)$
*   Use $S_0=S(t_0)$ to integrate the states up to $S(t_f)$ which approximates the solution to the differential equation

Example: Differential equation $\frac{df(t)}{dt}=e^-t$ using initial condition, $f_0=-1$, with an exact solution $f(t)=-e^-t$ and approximating between 0 and 2 with increments of 0.2.


```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda t, s: np.exp(-t)
h = 0.2
t = np.arange(0, 2+h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12,8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
